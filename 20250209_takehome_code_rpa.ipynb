# Decisions made for matching
In the hospital dataset, we have 3 plans:
1. Aetna Open Access Managed Choice
2. United Healthcare Choice Plus
3. Cigna Corporation National OAP

To attempt a match of payer & plan between a hospital dataset row and a payer dataset row, I first check for matching payer names. For Aetna and Cigna, this is just a case-insensitive match of 'Aetna' and 'Cigna' in the payer dataset. However, for United Healthcare this involves checking for both 'United' and 'UHC', since both are used in the payer dataset.

Next, ideally we would have a perfect plan name (or unique plan identifier) match between the two datasets. However, that is simply not the case, and has varying levels of success depending on the hospital and payer/plan. Additionally, we want to match the code type and number, and ideally we want to get the best match between negotiation_type in the hospital dataset and the combination of standard_charge_methodology and additional_generic_notes in the payer dataset.

I decided to consider classifying each of the three plans of interest somewhat categorically, since plan name matches are inexact, but plan types might match sufficiently. As mentioned earlier, all three plans seem to fall under the PPO umbrella, with the Cigna plan possible also including HMO. Therefore, for all three plans, I thought I would try to match plan name based on what I see or go with category if the option was available. If both category and plan name seemed to work, then there would be several options to consider.

The best solution I identified for right now, given time constraints: Create a list of acceptable plan names from the payer dataset, and match payer_name exactly (as described above) and then a plan name from the list. This list essentially includes the closest plan names and/or general labels that encompass PPO in some way (PPO, or commercial, or all plans). For payer dataset rows not matched with a hospital dataset row (i.e., not in the good_plan_names list), I will still include them in the output dataset, but noting clearly the lack of match.

As mentioned above, CPT code matching is easy, but ultimately we also need to match negotiation_type (in the payer dataset) to the combination of standard_charge_methodology and additional_generic_notes (in the hospital dataset).

```python
# Unique plan names across all 3 payers
payer_df[(payer_df.payer_name.str.contains('cigna', case=False))|(payer_df.payer_name.str.contains('aetna', case=False))|
         (payer_df.payer_name.str.contains('united', case=False))|(payer_df.payer_name.str.contains('uhc', case=False))].plan_name.unique().tolist()
```

```python
# Acceptable plan names, for now
good_plan_names = ['Commercial', 'All Payers', 'HMO and PPO', 'HMO/POS/PPO', 'cignappo1047', 'unitedhealthcareppo1160',
                   'cignaopenaccess1413', 'aetnappo1010', 'cignaopenaccessppo1465', 'aetnaopenaccesshmo1414']
# Dictionary linking each EID (key) to the license number for the same hospital (value)
hospital_matches = {131740114: '13-1740114', 
                    133971298: '7002053H', 
                    131624096: '330024'}
# Create an empty hospital dataset row to append to each payer dataset row without a match
empty_hospital = [None] * hospital_df.shape[1]
# Create an empty payer dataset row to append to each hospital dataset row without a match
empty_payer = [None] * payer_df.shape[1]
# Append "hospitaldf_" to each hospital dataset column header, for clarity
hospital_header = ['hospitaldf_' + str(col) for col in hospital_df.columns]
# Append "payerdf_" to each payer dataset column header, for clarity
payer_header = ['payerdf_' + str(col) for col in payer_df.columns]
# Combine the hospital and payer dataset headers
combined_header = hospital_header + payer_header
# Empty list to store matched rows (or unmatched rows, when such cases appear)
combined_df = []
# Empty list to store payers that have been matched to a hospital dataset row
payer_df_rows_matched = []

for index, row in hospital_df.iterrows():
    curr_payer = row.payer
    curr_ein = row.ein
    curr_license = hospital_matches[curr_ein]
    curr_code_type = row.code_type
    if curr_code_type == 'CPT':      # Convert CPT codes to strings
        curr_code = str(row.code)
    elif curr_code_type == 'MS-DRG': # Append MS-DRG to such codes, to match payer dataest format of these codes
        curr_code = 'MS-DRG '+ str(row.code)
    if curr_payer != 'unitedhealthcare': # Simple match for non-UHC payer
        matched_rows = payer_df[payer_df.payer_name.str.contains(curr_payer, case=False)]
    else: # Extra check for UHC payer
        matched_rows = payer_df[(payer_df.payer_name.str.contains('united', case=False))|
                                (payer_df.payer_name.str.contains('uhc', case=False))]
    matched_rows = matched_rows[matched_rows.license_number == curr_license]
    matched_rows = matched_rows[matched_rows.plan_name.isin(good_plan_names)]
    matched_rows = matched_rows[(matched_rows.code_type==curr_code_type)&(matched_rows.raw_code==curr_code)]
    # If at least one payer dataset row matches a hospital dataset row, combined them
    if matched_rows.shape[0] > 0:
        for payer_index, payer_row in matched_rows.iterrows():
            new_row = row.tolist() + payer_row.tolist()
            combined_df.append(new_row)
            payer_df_rows_matched.append(payer_index)
    # If no match, combine an empty payer row with that hospital dataset row
    else:
        new_row = row.tolist() + empty_payer
        combined_df.append(new_row)
        
payer_df_rows = sorted(list(set(payer_df_rows_matched)))
# If no hospital dataset row matches a payer dataset row, combine an empty hospital row with that payer dataset row
for index, row in payer_df.iterrows():
    if index not in payer_df_rows:
        new_row = empty_hospital + row.tolist()
        combined_df.append(new_row)
```

```python
# List of lists --> dataframe
combined_df = pd.DataFrame(combined_df, columns=combined_header)
combined_df.head()
```
